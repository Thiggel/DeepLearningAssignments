```python
import matplotlib.pyplot as plt
import numpy as np

lstm = dat[0:3]
ft_lstm = dat[3:6]
tree_lstm = dat[6:9]

lstm_mean = lstm.mean(0)
lstm_std = lstm.std(0)

ft_lstm_mean = ft_lstm.mean(0)
ft_lstm_std = ft_lstm.std(0)

tree_lstm_mean = tree_lstm.mean(0)
tree_lstm_std = tree_lstm.std(0)

models = ['LSTM', 'Fine-Tuned LSTM', 'Tree-LSTM']
datasets = 5

accuracies = np.array([lstm_mean, ft_lstm_mean, tree_lstm_mean])
std_dev = np.array([lstm_std, ft_lstm_std, tree_lstm_std])

plt.figure(figsize=(10, 6))

for i, model in enumerate(models):
    x = np.arange(1, datasets + 1)
    y = accuracies[i, :]
    error = std_dev[i, :]

    plt.plot(x, y, label=f'{model}', marker='o')

    plt.fill_between(x, y - error, y + error, alpha=0.1)

plt.xlabel('$n$th Fraction of Test Set', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.xticks(np.arange(1, datasets + 1), [f'{d}' for d in range(1, datasets + 1)])
plt.tick_params(axis='both', labelsize=18)
plt.legend(fontsize=16)
plt.tight_layout()
plt.grid(False)

plt.show()
```
